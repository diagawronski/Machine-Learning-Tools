## How to Upload and Use Data in Google CoLab





1) Import all Necessary packages

```python
import pandas as pd
from google.colab import drive
import matplotlib.pyplot as plt
```

2) Mount your Google Drive

```python
drive.mount('/content/gdrive', force_remount = True)
```

3) Pull the data from your Google Drive folder

```python
df = pd.read_csv('/content/gdrive/MyDrive/Econ 441B/sat_gpa.csv')
```

4) Plot your graph!

```python
plt.scatter(df['sat_total'], df['gpa_fy'])
plt.title("Students' High School SAT Scores and Freshmen Year College GPAs")
plt.ylabel("Freshmen Year College GPA")
plt.xlabel("HS SAT Score")
plt.show()
```

As shown above, there is a positive linear relationship between students' high school SAT scores and their freshmen year GPAs in college.
